```python
import json 
import pandas as pd
import numpy as np
```

```python
import nltk
from collections import defaultdict
import random as r
```

```python
with open('./masterchef_data_cleaned.json', 'r') as file:
    data = json.load(file)
data
```

```python
len(data['United States'])
```

```python
N = 3
total_ngrams = []
```

```python
country_list = ['United States', 'Canada', 'Brazil']
for country in country_list:
    for i in range(len(data[country])):
        # print(data['United States'][i]['evaluations'])
        ngrams_list = list(nltk.ngrams(data['United States'][i]['evaluations'],N))
        total_ngrams.append(ngrams_list)
```

```python
total_ngrams
```

```python
ngram_freq = defaultdict(int)
for ng_list in total_ngrams:
    for ng in ng_list:
        ngram_freq[ng] += 1
```

```python
ngram_freq
```

```python
def predict_next_word(previous):
    # for (ngram, freq) in ngram_freq.items():
    #     print(list(ngram)[:-1])
    #     print(ngram[0])
    idx = 1
    if len(previous) != 1:
        idx = len(previous)
    matching_ngrams = [(ngram, freq) for ngram, freq in ngram_freq.items() if list(ngram)[:-1] == previous[:idx]]
    if not matching_ngrams:
        return "Prediction Unavailable"
    sorted_ngrams = sorted(matching_ngrams, key=lambda x: x[1], reverse=True)
    print(sorted_ngrams)
    prediction = sorted_ngrams[0][0][-1]
    return prediction
```

```python
prev = ['WIN_T', 'WIN_I']
predict_next_word(prev)
```
